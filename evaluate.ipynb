Esta función guarda todos los nombres de los audios en nuestro csv a partir de una carpeta donde esten los audios

```python
def construir_csv(carpeta:str):
    carpeta_audios = carpeta
    archivos = os.listdir(carpeta_audios)

    audios = [f for f in archivos if os.path.isfile(os.path.join(carpeta_audios, f))]

    with open("data_adultos.csv", "w", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["AUDIO"])  
        for audio in audios:
            writer.writerow([audio])
```

Realizan las transcripciones y se calculan las métricas

```python
def eval(data_archivo:str, results, audio_root: str):
    model = whisper.load_model("turbo")

    carpeta_audios = Path(audio_root)
    print("LA CARPETA ES: ", carpeta_audios)
    if not carpeta_audios.exists():
        raise FileNotFoundError(f"Carpeta de audios no encontrada: {audio_root}")

    with open(data_archivo, encoding="utf-8") as f:
        r = csv.DictReader(f, delimiter=",")
        
        for registro in r: 
            audio = registro['AUDIO']
            audio_path = carpeta_audios / audio
            output = model.transcribe(str(audio_path), language="es", fp16=False)
            hyp = output["text"]


            results[audio] = dict()

            results[audio]["Transcripción"] = hyp
            #results[audio]["métricas"] = #calcular todas las métricas que tenemos.

```
